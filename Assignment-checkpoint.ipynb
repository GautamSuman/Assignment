```python
import pandas as pd
import numpy as np
```

```python
#reading CSV file
df = pd.read_csv("C:\\Users\\vinay\\Desktop\\SumanResume\\Assignment\\googleplaystore.csv")
```

```python
df.tail()
```

```python
#replacing "Varies with device" in size column with zero
df['Size'] = df['Size'].replace('Varies with device',0)
df.tail()
```

```python
#we have created new column "unit", where we are storing unit of size of app(mb/kb)
df['Unit'] = df['Size'].str.strip().str[-1]
df.tail(50)
```

```python
#here we are removing last character of size column, so that we can get integer value
df['Size'] = df['Size'].str.replace(r'\D', '')
df.tail(50)
```

```python
#here we are converting data type of size column to float, so that we can do mathematical operation over it
df["Size"] = df.Size.astype(float)
#df["Price"] = df.Size.astype(int)
df.dtypes
```

```python
#converting kb to mb 
df['Size'] = np.where(df['Unit'] =='k', df['Size']/1024,df['Size'])
```

```python
df.tail(50)
```

```python
#All values in size column are in mb so we have replaced all value in unit column to M in ordered to ovoid confusion
df['Unit'] = 'M'
df.tail(50)
```

2)What approach you will implement to fill the missing values of the
Size column & why?

ANS : There are many missing values in size column . So I will fill missing values by taking Median of whole Size column. I am taking median  because median  will not get effected by Outliers value ( Unlike mean which is very easily get affetc by outliers ) . Median is 33  

```python
#calculating meadian
Median= df['Size'].median()
```

```python
#fill meadian in the place of missing value
df['Size'].fillna(Median)
df.head(5)
```

3)What is the app name having max price?

Ans : "I'm Rich - Trump Edition" is app having Max price in given data . Implementation below .

```python
df['Price'] = df['Price'].str.replace('$', '') #removing $ from price column
df['Price'] = pd.to_numeric(df['Price'], errors='coerce') #removing string value from price column
df['Price'].replace('NaN',0) #replacing Nan with zero
```

```python
#logic to display max price and app name 
df[['App','Price']][df.Price == df['Price'].max()]
```

4. List top 20 apps in Game Category(Based on Number of Installation)

```python
df['Installs'] = df['Installs'].str.replace(r'\D', '') #removing last character of install column(plus sign)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce') #removing string value from install column
df['Installs'].replace('NaN',0) #replacing Nan with zero
```

```python
df.nlargest(20, ['Installs']) # finding top 20 largest value of install column with corresponding row
```

5)Is it better that developers use a concise word or words to name their app?
